```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
import warnings
import joblib

warnings.filterwarnings('ignore')


url = 'https://github.com/FlipRoboTechnologies/ML_-Datasets/blob/main/Temperature%20Forecast/temperature.csv?raw=true'
data = pd.read_csv(url)


print("First few rows of the dataset:")
print(data.head())


print("\nBasic information about the dataset:")
print(data.info())
print("\nSummary statistics of the dataset:")
print(data.describe())

missing_values = data.isnull().sum()
print("\nMissing values in the dataset:")
print(missing_values[missing_values > 0])

plt.figure(figsize=(10, 6))
sns.histplot(data['Next_Tmax'], bins=30, kde=True)
plt.title('Distribution of Next Day Maximum Temperature')
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(data['Next_Tmin'], bins=30, kde=True)
plt.title('Distribution of Next Day Minimum Temperature')
plt.show()

plt.figure(figsize=(14, 10))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

data.fillna(data.median(), inplace=True)


data.drop(['Date'], axis=1, inplace=True)


X_tmax = data.drop(['Next_Tmax', 'Next_Tmin'], axis=1)
y_tmax = data['Next_Tmax']
X_train_tmax, X_test_tmax, y_train_tmax, y_test_tmax = train_test_split(X_tmax, y_tmax, test_size=0.2, random_state=42)

X_tmin = data.drop(['Next_Tmax', 'Next_Tmin'], axis=1)
y_tmin = data['Next_Tmin']
X_train_tmin, X_test_tmin, y_train_tmin, y_test_tmin = train_test_split(X_tmin, y_tmin, test_size=0.2, random_state=42)

models = {
    'RandomForestRegressor': RandomForestRegressor(),
    'LinearRegression': LinearRegression(),
    'DecisionTreeRegressor': DecisionTreeRegressor()
}

for name, model in models.items():
    model.fit(X_train_tmax, y_train_tmax)
    y_pred_tmax = model.predict(X_test_tmax)
    print(f'{name} for Next_Tmax - Mean Squared Error:', mean_squared_error(y_test_tmax, y_pred_tmax))
    print(f'{name} for Next_Tmax - R2 Score:', r2_score(y_test_tmax, y_pred_tmax))

for name, model in models.items():
    model.fit(X_train_tmin, y_train_tmin)
    y_pred_tmin = model.predict(X_test_tmin)
    print(f'{name} for Next_Tmin - Mean Squared Error:', mean_squared_error(y_test_tmin, y_pred_tmin))
    print(f'{name} for Next_Tmin - R2 Score:', r2_score(y_test_tmin, y_pred_tmin))

param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

grid_tmax = GridSearchCV(RandomForestRegressor(), param_grid, cv=5, scoring='r2', n_jobs=-1)
grid_tmax.fit(X_train_tmax, y_train_tmax)
best_model_tmax = grid_tmax.best_estimator_

grid_tmin = GridSearchCV(RandomForestRegressor(), param_grid, cv=5, scoring='r2', n_jobs=-1)
grid_tmin.fit(X_train_tmin, y_train_tmin)
best_model_tmin = grid_tmin.best_estimator_

joblib.dump(best_model_tmax, 'best_model_tmax.pkl')
joblib.dump(best_model_tmin, 'best_model_tmin.pkl')

print("Models saved successfully.")
```
